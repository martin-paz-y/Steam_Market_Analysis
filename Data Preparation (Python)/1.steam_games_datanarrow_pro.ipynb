```python
import pandas as pd

# Display settings to show all columns and improve readability
pd.set_option('display.max_columns', None)      # Remove column limit
pd.set_option('display.max_rows', 50)           # Show up to 50 rows
pd.set_option('display.expand_frame_repr', False)  # Prevent line wrapping

# 1. Define the file path
path = r'C:\Users\MartínPazYáñez\Desktop\iron\bootcamp\TFM\steam_games_dataset\games.json'

# 2. Load the JSON file into a DataFrame
df = pd.read_json(path)

# 3. TRANSPOSE: rotate the DataFrame so fields like 'name', 'price', etc. become columns
df = df.T


df.head()
```

```python
# 2. Create a real engagement metric
df['total_ratings'] = df['positive'].fillna(0) + df['negative'].fillna(0)

# 3. Filter games WITH engagement (games that have ratings)
df = df[df['total_ratings'] > 0]

# 4. Convert release_date to datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# 5. Filter last 8–9 years (2016 onwards to focus on more recent data)
df = df[df['release_date'] >= '2016-01-01']

# 6. Remove rows missing critical data
columnas_criticas = ['name', 'price', 'developers', 'publishers', 'categories', 'genres']
df = df.dropna(subset=columnas_criticas, how='any')

# 7. Filter games with minimum engagement (e.g. at least 20 total ratings)
df = df[df['total_ratings'] >= 20]

# 8. If the dataset is still larger than 20k rows, keep the most relevant games by engagement
if len(df) > 19500:
    df = df.nlargest(20000, 'total_ratings')

# 9. Save the filtered dataset
df.to_csv('games_filtered.csv', index=True)

df.info
```
