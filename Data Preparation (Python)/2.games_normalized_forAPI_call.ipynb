```python
import pandas as pd

# Display settings to show all columns and improve readability
pd.set_option('display.max_columns', None)        # Remove column limit
pd.set_option('display.max_rows', 50)             # Show up to 50 rows
pd.set_option('display.expand_frame_repr', False) # Prevent line wrapping

# 1. Define the file path
path = r'C:\Users\MartínPazYáñez\Desktop\iron\bootcamp\TFM\steam_games_dataset\steam_games_datanarrow_pro.csv'

# 2. Load the CSV file into a DataFrame
df_steam= pd.read_csv(path)

# 3. Show the first rows of the DataFrame
df_steam = df_steam.rename(columns={'Unnamed: 0': 'app_id'})

df_steam.head()
```

```python
# RAWG API uses a text-based "search" parameter with fuzzy matching by default.
# However, game names from Steam often contain extra symbols, editions, or formatting
# (e.g. ™, ®, colons, dashes, or special characters) that can reduce search accuracy.
# Normalizing the game name helps improve matching reliability when calling the RAWG API,
# especially when using search_precise or search_exact options.
```

```python
def normalize_game_name(name):
    """Normalize game name for RAWG search"""
    if pd.isna(name):
        return name
    
    # Remove common symbols
    name = name.replace('™', '')
    name = name.replace('®', '')
    name = name.replace(':', '')
    name = name.replace('–', '-')
    name = name.replace('—', '-')
    
    # Remove extra spaces
    name = ' '.join(name.split())
    
    return name.strip()

# Find name position FIRST (before creating the column)
name_position = df_steam.columns.get_loc('name')

# Insert name_normalized right after 'name'
df_steam.insert(name_position + 1, 'name_normalized', df_steam['name'].apply(normalize_game_name))

# Save the file with the new column
df_steam.to_csv('games_normalized_forAPI_call.csv', index=False)

# Show some examples
print(df_steam[['name', 'name_normalized']].head(10))
```
